```python
#1. Analyse and improve code from data structures lab with loop
# - Parts of code to review and improve: collecting inventory quantity & customer orders

# - Inventory quantity already uses a loop but can be improved as below to ensure input is a valid number (non-negative + non-string) and customer orders already used a loop but can be improved by making sure the input is exact three orders.

# Initialize products list
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Create empty dictionary for the inventory
inventory = {}

# Loping through each product to get the inventory quantity
for product in products:
    quantity = -1  # Starting with an invalid number for validation

    while quantity < 0:  # Ensuring the loop runs until a valid number is entered
        try:
            quantity = int(input(f"Please enter a non-negative number for {product}: "))
            if quantity < 0:
                print("Quantity must be a non-negative number. Please try again.")

        except ValueError: # Handle the case where the input is not a valid integer
            print("Invalid input. Please enter a number.")

    inventory[product] = quantity  # Storing the valid quantity in the inventory dictionary

print("Inventory:")
print(inventory)

customer_orders = set()

for _ in range(3):  # Loop exactly three times
    order = input(f"Please enter a product to order from {products}: ")

    # Validation: Check if the entered order is valid (exists in products)
    if order in products:
        customer_orders.add(order)
    else:
        print("Invalid product. Please enter a exact product name.")

# Message to user after the loop 
print("Thank you for your order! You've selected the following products:")
for product in customer_orders:
    print("-", product)
```

```python
# Answer q2

products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {}

for product in products:
    quantity = -1 

    while quantity < 0:
        try:
            quantity = int(input(f"Please enter a non-negative number for {product}: "))
            if quantity < 0:
                print("Quantity must be a non-negative number. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    inventory[product] = quantity

print("Inventory:")
print(inventory)

customer_orders = set()

while True:
    order = input(f"Please enter a product to order from {products}: ")

    if order in products:
        customer_orders.add(order)
    else:
        print("Invalid product. Please enter a valid product name.")

    continue_ordering = input("Do you want to add another product? (Yes/y or No/n): ").strip().lower()

    acceptable_yes = {"yes", "y"}
    acceptable_no = {"no", "n"}

    if continue_ordering in acceptable_no:
        break
    elif continue_ordering not in acceptable_yes:
        print("Invalid input. Please respond with 'y' or 'n'.")

print("Thank you for your order! You've selected the following products:")
for product in customer_orders:
    print("-", product)
```

```python
# Answer q3

products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {
    "t-shirt": 10,
    "mug": 20,
    "hat": 15,
    "book": 8,
    "keychain": 5
}

customer_orders = {"t-shirt", "mug", "hat"}

for order in customer_orders:
    if inventory[order] > 0:
        inventory[order] -= 1
    else:
        print(f"Inventory for {order} is already at zero.")
        
print("Updated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
