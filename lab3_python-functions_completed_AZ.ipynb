```python
# Answer q1: Define initialize_inventory function by taking products as parameter and initialize inventory by using loop & user input 

def initialize_inventory(products):
    inventory={}
    for product in products:
        while quantity < 0:
            try:
                quantity = int(input(f"Please enter a non-negative number for {product}: "))
                if quantity < 0:
                    print("Quantity must be a non-negative number. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a number.")
        inventory[product] = quantity
    return inventory
```

```python
# Answer q2: Define get_customer_orders function update_inventory taking custom orders and inventory as parameters

def get_customer_orders():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    customer_orders = set()

    while True:
        order = input(f"Please enter a product to order from {products}: ")
        if order in products:
            customer_orders.add(order)
        else:
            print("Invalid product. Please enter a valid product name.")
        
        continue_ordering = input("Do you want to add another product? (Yes/y or No/n): ").strip().lower()
        if continue_ordering in {"no", "n"}:
            break
        elif continue_ordering not in {"yes", "y"}:
            print("Invalid input. Please respond with 'y' or 'n'.")
    
    return customer_orders

customer_orders = get_customer_orders()
```

```python
# Answer q3: Define update_inventory function with customer orders and inventory as parameters

def update_inventory(customer_orders, inventory):
    for order in customer_orders:
        if inventory.get(order, 0) > 0:
            inventory[order] -= 1
        else:
            print(f"Inventory for {order} is already at zero.")
```

```python
# Answer q4: Define a function named `calculate_order_statistics` function with customer_orders and products as parameters

products = ["t-shirt", "mug", "hat", "book", "keychain"]

def calculate_order_statistics(customer_orders, products): #Total products ordered is size of the customer_orders set
    total_products_ordered = len(customer_orders)
    
    if len(products) > 0: 
        percentage_unique_products_ordered = (total_products_ordered / len(products)) * 100
    else:
        percentage_unique_products_ordered = 0  # Avoiding division by zero

    return total_products_ordered, percentage_unique_products_ordered # Returning both statistics as a tuple

total_products_ordered, percentage_unique_products_ordered = calculate_order_statistics(customer_orders, products)

print(total_products_ordered, percentage_unique_products_ordered)
```

```python
# Answer q5: Define a function named print_order_statistics
   
def print_order_statistics(order_statistics):
    
    total_products_ordered, percentage_unique_products_ordered = order_statistics
    
    print(f"Total unique products ordered: {total_products_ordered}")
    print(f"Percentage of unique products ordered: {percentage_unique_products_ordered:.2f}%")
```

```python
# Answer q6: Define a function named print_updated_inventory that takes inventory as a parameter

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
```
